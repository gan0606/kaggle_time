```python
# 基本moduleのimport
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 基本moduleの設定module
plt.rcParams['font.family'] = "MS Gothic"
plt.style.use("ggplot")
sns.set_theme(context="talk", style="darkgrid", font="MS GOTHIC")
```

```python
# 出力行数の変更
pd.set_option("display.max_rows", 100)
```

```python
# データの読み込み
train_df = pd.read_csv("processed/train_pr.csv")
test_df = pd.read_csv("processed/test_pr.csv")
submission_df = pd.read_csv("datasets/sample_submission.csv")
```

```python
# 時系列データなのでMinMaxScalerを使う
from sklearn.preprocessing import MinMaxScaler
```

```python
# インスタンスを生成
mms = MinMaxScaler()
```

```python
target = train_df["sales"]
```

```python
# train_dfとtest_dfを結合
all_df = pd.concat([train_df.drop("sales", axis=1), test_df], axis=0)
all_df
```

```python
# scalingするカラム
select_columns = ['family', 'onpromotion', 'year', 'month', 'day_of_week', 'day', 'is_holidays', 'oil_price', 'transactions']
```

```python
all_df[select_columns]
```

```python
scaling_data = mms.fit_transform(all_df[select_columns])
```

```python
scaled_all_df = pd.DataFrame(scaling_data, columns=mms.get_feature_names_out())
scaled_all_df
```

```python
id_date = all_df[["id", "date"]]
id_date
```

```python
scaled_all_df_new = pd.concat([id_date.reset_index(drop=True), scaled_all_df.reset_index(drop=True)], axis=1)
scaled_all_df_new
```

```python
train_df_new = scaled_all_df_new[:len(train_df)]
train_df_new = pd.concat([train_df_new, target], axis=1)
train_df_new
```

```python
test_df_new = scaled_all_df_new[len(train_df):].reset_index(drop=True)
test_df_new
```

```python
# 保存
train_df_new.to_csv("processed/train_pr_sc.csv", index=False)
test_df_new.to_csv("processed/test_pr_sc.csv", index=False)
```
